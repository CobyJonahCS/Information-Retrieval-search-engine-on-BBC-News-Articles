```python
from nltk.tokenize import word_tokenize

df = pd.read_csv(r"C:/Users/charl/Documents/IRProject/Information-Retrieval-search-engine-on-BBC-News-Articles/data/bbc-news-data.csv", on_bad_lines='skip')
```

```python
stop_words = set(stopwords.words("english"))
stemmer = PorterStemmer()
```

```python
def preprocess_text(text, use_stemming=False):
    if pd.isna(text):
        return ""

    # Convert to lowercase
    text = text.lower()

    # Remove punctuation and numbers
    text = re.sub(r"[^a-z\s]", " ", text)

    # Tokenise
    tokens = word_tokenize(text)

    # Remove stopwords and very short tokens
    tokens = [token for token in tokens if token not in stop_words and len(token) > 1]

    # Optional stemming
    if use_stemming:
        tokens = [stemmer.stem(token) for token in tokens]

    # Return cleaned text as a single string
    return " ".join(tokens)
```

```python
df["processed_text"] = df["text"].apply(lambda x: preprocess_text(x, use_stemming=False))

# Optional version for experiments
df["processed_text_stemmed"] = df["text"].apply(lambda x: preprocess_text(x, use_stemming=True))

# -----------------------------
# 5. Keep relevant columns
# -----------------------------
df = df[["processed_text", "processed_text_stemmed", "labels"]]

# -----------------------------
# 6. Save processed dataset
# -----------------------------
df.to_csv("data/bbc_articles_processed.csv", index=False)

print("✅ Preprocessing complete. File saved as bbc_articles_processed.csv")
```
